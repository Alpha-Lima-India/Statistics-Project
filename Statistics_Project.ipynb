```r
# part a Q1
flag_dots = rep(1, times = 100000)
#print(flag_dots)

bernoli_choosing = rbinom(n = 100000, size = 1, prob = 0.2)
#print(bernoli_choosing)

planted_flags = flag_dots * bernoli_choosing
distance_between_ones = which(planted_flags == 1)


# store the distance between two neighbor flags in a vector
distances = diff(distance_between_ones / 100000)
#cat(distances)
#cat(planted_flags)

library(ggplot2)

hist(distances, breaks = 80, col = "skyblue", xlab = "Distance", ylab = "Frequency", main = "Distribution of Flag Distances with built-in R functions")
ggplot(data = data.frame(distances), aes(x = distances)) + geom_histogram(binwidth = 0.00001, fill = "skyblue", color = "navyblue") + xlab("Distance") + ylab("Frequency") + ggtitle("Distribution of Flag Distances with ggplot")
#ggplot(data = data.frame(distances), aes(x = distances)) + geom_histogram(binwidth = 0.00001, fill = "skyblue", color = "navyblue") + xlab("Distance") + ylab("Frequency") + ggtitle("Distances Logarithmic Histogram") + coord_cartesian(ylim = c(0, max(hist(distances)$counts))) + scale_y_continuous(trans = "log10")

```

```r
#part b Q1

random_numbers <- runif(55000)
final_vector = c()
for (i in random_numbers) {
  if (length(final_vector) < 50000) {
    if (!(i %in% final_vector)) {
    final_vector = c(final_vector,i)
  }
  }
}
planed_flags_loc = sort(final_vector)
# distance between two neighbor location
flags_distances <- planed_flags_loc[-1] - planed_flags_loc[-length(planed_flags_loc)]
hist(flags_distances, breaks = 100, col = "skyblue", xlab = "Distance", ylab = "Frequency", main = "Distribution of Flag Distances with built-in R functions")
ggplot(data = data.frame(flags_distances), aes(x = flags_distances)) + geom_histogram(binwidth = 0.000001, fill = "skyblue", color = "navyblue") + xlab("Distance") + ylab("Frequency") + ggtitle("Distribution of Flag Distances with ggplot")

```

```r
# Step 1: Load the required library
library(ggplot2)

# Step 2: Define variables and parameters
H <- 1
n <- 100000
p <- 0.2

# Step 3: Simulate descent movement
distances <- rexp(n, rate = 1/p) # Generate random distances based on exponential distribution

# Step 4: Calculate flag positions
flags <- cumsum(distances) # Calculate cumulative sum of distances for flag positions

# Step 5: Calculate distances between consecutive flags
consecutive_distances <- diff(flags) # Calculate differences between consecutive flags

# Step 6: Filter out zero and negative values
consecutive_distances <- consecutive_distances[consecutive_distances > 0]

# Step 7: Check for problematic values
problematic_values <- which(is.infinite(log10(consecutive_distances)))

# Step 8: Print problematic values
print(problematic_values)

# Step 9: Create a ggplot histogram with logarithmic scale
ggplot(data = data.frame(consecutive_distances), aes(x = consecutive_distances)) +
  geom_histogram(binwidth = 0.1, fill = "skyblue", color = "black") +
  xlab("Distance between Consecutive Flags") +
  ylab("Frequency (log scale)") +
  ggtitle("Distribution of Flag Distances (Logarithmic Scale)") +
  scale_y_log10()
```
